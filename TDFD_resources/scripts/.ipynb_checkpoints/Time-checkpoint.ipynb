# IQ demoulation of a single subcarrier by definition
IQ demodulation is widely used in digital communication and signal processing due to its spectrum efficiency. IQ modulation uses two components, the in-phase(I) component and the quadrature(Q) component (which is $90^\circ$ shifted from the I component).

If a time-varying complex signal is given as
$$
S(t) = a(t)+\mathrm{j}b(t),
$$

which is transmitted on a $\cos(\omega t)$ carrier, then the modulated carrier is found to be
$$
C(t)  = [a(t)+\mathrm{j}b(t)]\cos(\omega t).
$$

At the receiver end, the original signal $S(t)$ can be extracted with IQ demodulation. The I component can be resolved by
$$
I(t) = Re[LF \left( C(t)\times \cos(\omega t) \right)] = \frac{a(t)}{2}\\
Q(t) = Re[ LF \left( C(t)\times \sin(\omega t) \right)] = -\frac{b(t)}{2}
$$
Thus we have
$$
S(t) = 2 Conj[ I(t) +\mathrm{j}Q(t)]
$$

## IQ modulation and QPSK mapping of the subcarrier

## IQ demodulation

A direct way to demodualte the input signal can be concluded by the definition of IQ modulation.


```python
from collections import deque

def RollingAverage(array, window_width):
    currentSum = 0
    result = []
    queue = deque([0]*window_width, maxlen = window_width)
    for i in range(len(array)):
        currentSum -= queue.popleft()
        queue.append(array[i])
        currentSum += array[i]
        result.append(currentSum/window_width)
    return np.array(result)
        

def PlotPair(ts, sig1, sig2, label1, label2, istart = 300, istop = int(1e5)):
    plt.plot(ts[istart:istop]*1e6,sig1[istart:istop], label=label1)
    plt.plot(ts[istart:istop]*1e6,sig2[istart:istop], label=label2)
    plt.xlabel("Time (us)")
    plt.legend()
    plt.show()

def DemodulateIQ(sc, fc, tau):
    ref_i = np.cos(2*np.pi*fc*g_t_array)*2
    ref_q = np.sin(2*np.pi*fc*g_t_array)*2

    sc_demo_i = np.real(np.multiply(sc,ref_i))
    sc_demo_q = np.real(np.multiply(sc, ref_q))
    sc_result_i = RollingAverage(sc_demo_i, tau)
    sc_result_q = RollingAverage(sc_demo_q, tau)
    PlotPair(g_t_array, sc_result_i, sc_result_q, "Lowpass I", "Lowpass Q")    
    return (sc_result_i, sc_result_q)
    
resI, resQ = DemodulateIQ(sc, fc, g_slotSampleLen)
```

```python
def qpskDemod(cnum):
    return np.around((1- np.sqrt(2)*np.real(cnum))/2) + np.around(1-np.sqrt(2)*np.imag(cnum))

res_i_resample = [resI[(i+1)*g_slotSampleLen] for i in range(symbolNum)]
res_q_resample = [resQ[(i+1)*g_slotSampleLen] for i in range(symbolNum)]

res = [qpskDemod(res_i_resample[i]-1j*res_q_resample[i]) for i in range(symbolNum)]
plt.plot(res, label="Demodulated", drawstyle='steps-pre')
plt.plot(signal, label = "Original", drawstyle='steps-pre')
plt.legend()
plt.show()
```

```python
def qpskDemod(cnum):
    return np.around((1- np.sqrt(2)*np.real(cnum))/2) + np.around(1-np.sqrt(2)*np.imag(cnum))

def Demodulate_FFT(sc, fc_list, slot_num, tau, fs):
    freq_fft = np.arange(tau)/tau*fs
    freq_bitmap = [False]*tau
    for fc_idx in np.array(fc_list)/(fs/tau):
        freq_bitmap[int(fc_idx)] = True
    IQ = []
    for i in range(slot_num):
        sc_fft = (np.fft.fft(sc[i*tau:(i+1)*tau])/tau)[freq_bitmap]
        IQ.append(sc_fft)
     
    IQ = np.array(IQ)
    quant = qpskDemod(IQ)    
    return quant.T
```

```python
res = Demodulate_FFT(carrier, fc_set, symbolNum, g_slotSampleLen, g_fs)
```
